```python
def image2gray(image):
  # R G B 0.21 * red + 0.72 * green + 0.07 * blue
  image = image.permute(1,2,0)
  red  = image[:,:,0]
  green  = image[:,:,1]
  blue  = image[:,:,2]
  gray_image = 0.21 * red + 0.72 * green + 0.07 * blue
  plt.imshow(gray_image,  cmap='gray')
```
